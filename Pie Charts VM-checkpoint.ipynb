## Base Excel Merged File (Local File)

```python
# Load the Excel files
file1_path = "RequestFileV2.xlsx"
file2_path = "VM allocations.xlsx"

df1 = pd.read_excel(file1_path, sheet_name="Form responses 1")
df2 = pd.read_excel(file2_path, sheet_name="Production")

# Limit df2 to the first 108 rows
df2 = df2.iloc[:108]

# Clean "VM name" columns
df1["VM name"] = df1["VM name"].astype(str).str.strip()
df2["VM name"] = df2["VM name"].astype(str).str.strip()

# Drop rows where "VM name" is "nan"
df1 = df1[df1["VM name"].str.lower() != "nan"]
df2 = df2[df2["VM name"].str.lower() != "nan"]

# Merge DataFrames on "VM name" using an inner join to keep only matching records
merged_df = pd.merge(df1, df2, on="VM name", how="inner")

for col in merged_df.columns:
    if merged_df[col].dtype == "float64":  # Convert only numeric columns
        merged_df[col] = merged_df[col].astype(object)

# Replace empty strings with "Unknown"
merged_df.replace("", "Unknown", inplace=True)

# Fill NaN values with "Unknown"
merged_df.fillna("Unknown", inplace=True)
```

# Definition of Pie Charts Functions

## Use Precursor or Pathfinder Data

```python
def PrecusrorCount(df):
    # Precursor count for Google spreadsheet via Cloud
    column_name = 'Is your project using SKA precursor data?'  # Name of the column to analyze
    
    # Count non-Unknown cells in the column
    Precursor = df[df[column_name] != "Unknown"][column_name].count()    
    return(Precursor)
    print(Precursor)

def PathfinderCount(df):
    # Pathfinder count for Google spreadsheet via Cloud
    column_name = 'Is your project using SKA pathfinder data?'  # Name of the column to analyze

    # Count non-Unknown cells in the column
    Pathfinder = df[df[column_name] != "Unknown"][column_name].count()
    return(Pathfinder)
    print(Pathfinder)

def OtherInstCount(df):
    # Other instruments count for Google spreadsheet via Cloud
    column_name = 'Other Instruments used'  # Name of the column to analyze

    # Count non-Unknown cells in the column
    OtherInstrument = df[df[column_name] != "Unknown"][column_name].count()
    return(OtherInstrument)
    print(OtherInstrument)

def DataArchSimCount(df):
    # Granted data, Archive data or simulations count for Google spreadsheet via Cloud
    column_name = 'Is the project using granted data, archive data or simulations?'  # Name of the column to analyze

    # Count non-Unknown cells in the column
    DataArchiveSimulations = df[df[column_name] != "Unknown"][column_name].count()    
    return(DataArchiveSimulations)
    print(DataArchiveSimulations)

def POPDataPC(df):

    # Data Acquisition
    Precursor = PrecusrorCount(df)
    Pathfinder = PathfinderCount(df)
    NoPoP = DataArchSimCount(df) + OtherInstCount(df)

    total_count = Precursor + Pathfinder + NoPoP
    print(f"Sum of Precursor, Pathfinder, and NoPoP: {total_count}")

    
    # Data for the piechart
    labels = ['Precursor', 'Pathfinder', 'No']
    sizes = [Precursor, Pathfinder, NoPoP]
    
    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title('Use Precursor or Pathfinder Data ?')
    
    # Show the chart
    plt.show()
```

## Archival or Granted Data

```python
def DoGDADoS(df):
    
    # Selecting the column of interest
    column_name = 'Is the project using granted data, archive data or simulations?'
    
    # Replace empty strings with NaN
    df[column_name] = df[column_name].replace("", pd.NA)
    
    # Count all the different values
    Distribgdads = df[column_name].value_counts()
    
    # Define the groups to combine
    Distribgdads['Data Granted'] = Distribgdads.get('Granted', 0) + Distribgdads.get('Granted data', 0) + Distribgdads.get('Granted data, Archive data', 0)
    Distribgdads['Archived Granted'] = Distribgdads.get('Archive data', 0) + Distribgdads.get('Archival, simulations', 0) + Distribgdads.get('Granted data, Archive data', 0)
    Distribgdads['Simulations'] = Distribgdads.get('Archival, simulations', 0)
    # Combine 'Unknown' and 'Unkown'
    Distribgdads['Unknown'] = Distribgdads.get('Unknown', 0) + Distribgdads.get('Unkown', 0)
    # Drop the original categories after merging
    Distribgdads = Distribgdads.drop(['Granted', 'Granted data','Archive data','Archival, simulations','Granted data, Archive data', 'Unkown'], errors='ignore')
        
    # Data for the pie chart
    labels_combined = Distribgdads.index
    sizes_combined = Distribgdads.values
    
    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(sizes_combined, labels=labels_combined, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title('Distribution of Granted Data, Archive Data or Simulation')
    plt.show()
```

## Instruments used

```python
def IU(df):
    # List of columns to analyze
    columns_to_analyze = [
        'Is your project using SKA precursor data?',
        'Is your project using SKA pathfinder data?',
        'Other Instruments used',
        'Is the project using granted data, archive data or simulations?'
    ]
    
    # Initialize an empty series to hold combined counts
    InstrumetUsed_combined = pd.Series(dtype=int)
    
    # Loop through each column and count values
    for column in columns_to_analyze:
        if column in df.columns:
            # Replace empty strings with NaN for the column being processed
            df[column] = df[column].replace("", pd.NA)            
            # Count unique values in the column
            InstrumentUsed = df[column].value_counts()
            # Add the counts to the combined series
            InstrumetUsed_combined = InstrumetUsed_combined.add(InstrumentUsed, fill_value=0)
        else:
            print(f"Column '{column}' not found in the dataset.")
    
    # Reorganisation of the datas
    InstrumetUsed_combined['Others (Simulations, ...)'] = InstrumetUsed_combined.get('Archival, simulations', 0) + InstrumetUsed_combined.get('Granted', 0) + InstrumetUsed_combined.get('Granted data', 0) + InstrumetUsed_combined.get('Granted data, Archive data', 0) + InstrumetUsed_combined.get('Archive data', 0) + InstrumetUsed_combined.get('Unkown', 0)
    InstrumetUsed_combined['MeerKat'] = InstrumetUsed_combined.get('MeerKAT', 0) + InstrumetUsed_combined.get('Meerkat', 0)
    InstrumetUsed_combined = InstrumetUsed_combined.drop(['Granted', 'Granted data','Archive data','Archival, simulations','Granted data, Archive data','Unkown','MeerKAT','Meerkat'], errors='ignore')
    
    
    # Data for the pie chart
    labels = InstrumetUsed_combined.index
    sizes = InstrumetUsed_combined.values
    
    # Create figure and plot
    plt.figure(figsize=(8, 8))
    wedges, texts = plt.pie(sizes, labels=None, autopct=None, startangle=90, colors=plt.cm.Paired.colors)
    
    # Add a legend with both labels and percentages
    legend_labels = [f"{label}: {size:.1f}%" for label, size in zip(labels, sizes / sum(sizes) * 100)]
    plt.legend(wedges, legend_labels, loc="center left", bbox_to_anchor=(1, 0.5), title="Instrument Categories")
    
    # Adjust the title positioning
    plt.title('Instrument Used')
    
    # Show plot
    plt.show()
```

## Center Affiliated

```python
def CenterAffiliated(df):
    # Selecting the column of interest
    column_name = 'Affiliation'

    # Replace empty strings with NaN
    df[column_name] = df[column_name].replace("", pd.NA)
    
    # Count all the different values
    Affiliation = df[column_name].value_counts()
        
    # Reorganisation of data
    Affiliation['IAA'] = Affiliation.get('Instituto de Astrofísica de Andalucía (IAA)', 0) + Affiliation.get('IAA', 0) + Affiliation.get('Instituto de Astofísica de Andalucia (IAA-CSIC)', 0) + Affiliation.get('IAA-CSIC', 0)
    Affiliation = Affiliation.drop(['Instituto de Astrofísica de Andalucía (IAA)','Instituto de Astofísica de Andalucia (IAA-CSIC)','IAA-CSIC'], errors='ignore')
    
    # Data for the pie chart
    labels_combined = Affiliation.index
    sizes_combined = Affiliation.values
    
    # Create the pie chart
    plt.figure(figsize=(8, 8))
    wedges, texts = plt.pie(
        sizes_combined, labels=None, autopct=None, startangle=90, 
        colors=plt.cm.Paired.colors
    )
    
    # Ajouter une légende avec les labels et les pourcentages
    legend_labels = [f"{label}: {size:.1f}%" for label, size in zip(labels_combined, sizes_combined)]
    plt.legend(wedges, legend_labels, loc="center left", bbox_to_anchor=(1, 0.5), title="Legend")
    
    # Éloigner le titre
    plt.title('Center Affiliated')
    
    plt.show()
```

## Associated IAA Department

```python
def AssoIAADep(df):
    # Selecting the column of interest
    column_name = 'Department'
    
    # Replace empty strings with NaN
    df[column_name] = df[column_name].replace("", pd.NA)
    
    # Count all the different values
    Department = df[column_name].value_counts()
    
    # Reorganisation of data
    Department['Extragalactic Aastronomy'] = Department.get('Extragalactic Aastronomy', 0) + Department.get('Astronomía extragaláctica', 0) + Department.get('Extragalactic Astronomy', 0)
    Department['Radio Astronomy and Galactic Structure'] = Department.get('Radio Astronomy and Galactic Structure', 0) + Department.get('Radioastronomía y estructura galáctica', 0)
    Department = Department.drop(['Astronomía extragaláctica','Extragalactic Astronomy','Radioastronomía y estructura galáctica'], errors='ignore')
    
    # Data for the pie chart
    labels_combined = Department.index
    sizes_combined = Department.values
    
    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(sizes_combined, labels=labels_combined, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title('Associated IAA Department')
    plt.show()
```

```python
def MainScTop(df):
    # Define the column of interest
    column_name = 'Main scientific topic'
    
    # Identify the first column with the correct name
    matching_columns = [col for col in df.columns if col == column_name]
    if not matching_columns:
        raise ValueError(f"The column '{column_name}' is not present in the DataFrame.")
    
    # Select the first occurrence of the column
    first_column_index = df.columns.get_loc(matching_columns[0])
    df = df.iloc[:, [first_column_index]]

    # Replace empty strings with NaN
    df[column_name] = df[column_name].replace("", pd.NA)

    # Drop rows with NaN values
    df = df.dropna(subset=[column_name])
    print(df)

    # Count all the different values
    scientific_topic = df[column_name].value_counts()

    # Data for the pie chart
    labels_combined = scientific_topic.index
    sizes_combined = scientific_topic.values

    # Create the pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(sizes_combined, labels=labels_combined, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title('Main Scientific Topic of Each Project')
    plt.show()
```
